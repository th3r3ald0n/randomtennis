```python
import scipy
import math 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import randomtennis
```

# Explore game, set and match win probability based on point win probability
The point win probability stays constant during the match 

```python
game_prob, set_prob, match_prob_3, match_prob_5 = [], [], [], []
for p in np.arange(0.5, 0.7,0.01):
    game_prob.append(randomtennis.win_game_prob(p))
    #
    test = randomtennis.Simple_tester(p, winning_sets = 2)
    test.multiple_runs(10000)
    match_prob_3.append(test.player1.matches / (test.player1.matches + test.player2.matches))
    set_prob.append(test.player1.sets_total / (test.player1.sets_total + test.player2.sets_total))
    
    test2 = randomtennis.Simple_tester(p, winning_sets = 3)
    test2.multiple_runs(10000)
    match_prob_5.append(test2.player1.matches / (test2.player1.matches + test2.player2.matches))
```

```python
fig, ax = plt.subplots()
ax.plot(np.arange(0.5,0.7,0.01), match_prob_5, label = "Match (best-of-5)")
ax.plot(np.arange(0.5,0.7,0.01), match_prob_3, label = "Match (best-of-3)")
ax.plot(np.arange(0.5,0.7,0.01), set_prob, label = "Set")
ax.plot(np.arange(0.5,0.7,0.01), game_prob, label = "Game")
#ax.axis("equal")
ax.set(xticks = np.arange(0.5,0.7,0.025))
ax.tick_params(axis="x", labelrotation =90)
ax.legend()
ax.grid()
ax.set_xlim(left = 0.5, right = 1)
ax.set_ylabel("Winning probability")
ax.set_xlabel("Point winning probability")
ax.set_title("Winning probabilities")
```

# Exploring matches (best-of-five) with alternating win probabilities

```python
s_vs_s = {}
for i in np.arange(0.4, 0.8, 0.025):
    tmp = []
    for j in np.arange(0.4, 0.8, 0.025):
        test = randomtennis.Service_tester(i, j, winning_sets = 3)
        test.multiple_runs(10000)
        tmp.append(test.player1.matches / (test.player1.matches + test.player2.matches))
    s_vs_s[i] = tmp
```

## How to read the table
Player 1 wins 65% of points during his serve. Player2 wins 55% of points during his serve. The probability of player 1 winning the match (best-of-5) is about 94%

Note: The values on the diagonal are constant

```python
df = pd.DataFrame.from_dict(s_vs_s)
df.index = np.arange(0.4, 0.8, 0.025)
df
```

```python
mask = np.ones(df.shape,dtype='bool')
mask[np.triu_indices(len(df))] = False
fig, ax = plt.subplots()
sns.heatmap(df, xticklabels = np.round(df.columns,3), yticklabels = np.round(df.index,3), mask = mask, ax = ax)
ax.set_xlabel("Player 1's probability of winning a point during his service")
ax.set_ylabel("Player 2's probability of winning a point during his service")
ax.set_title("Player 1's probability of winning the match")
```

```python
#Plot the first line of the dataframe
fig, ax = plt.subplots()
ax.plot(np.round(df.iloc[0].index -0.4, 3), df.iloc[0].values, marker = "x")
ax.grid()
ax.set_title("Probability to win a match (values on the diagonals of heatmap)")
ax.set_ylabel("Probability to win match")
ax.set_xlabel("Difference in probabilities to win a point during own service game")
```
